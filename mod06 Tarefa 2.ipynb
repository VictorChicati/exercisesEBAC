```python
# Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
df = pd.read_csv('SINASC_RO_2019.csv') # Carregando a base de dados
```

```python
df.columns # visualizando as colunas da base de dados
```

```python
sns.boxplot(y='IDADEMAE', palette='pastel', data=df) # Plotando o gráfico de caixa para a variável idade da mãe
```

####  2) Faça um boxplot do peso do bebê

```python
sns.boxplot(y='PESO', data=df, palette='pastel') #utilizando a ferramenta do seaborn
```

####  3) Faça um histograma da idade da mãe

```python
sns.histplot(x='IDADEMAE', data=df, kde=True, alpha = .15) #utilizando a ferramenta do seaborn construir o gráfico
```

#### 4) Faça um histograma do peso do bebê

```python
sns.histplot(x='PESO', data=df, kde=True, alpha = .15) #utilizando a ferramenta do seaborn para construir o gráfico
```
