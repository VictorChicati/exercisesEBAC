````python
# importando todos os pacotes
import pandas as pd

import seaborn as sns
from seaborn import load_dataset

import matplotlib.pyplot as plt

import numpy as np

import statsmodels.api as sm
import statsmodels.formula.api as smf

from sklearn.metrics import r2_score
````

### 1. Explorando a base de gorjetas

a. Na aula fizemos uma regressão de ```tip``` explicada por ```net_bill```. Carregue os pacotes e a base, crie as variáveis necessárias como na aula, e reproduza esta regressão.

b. Reproduza também o gráfico de dispersão dessas duas variáveis.

````python
tips = sns.load_dataset('tips')
tips['tip_pct'] = (tips['tip']/ (tips['total_bill']))*100
tips['net_bill'] = tips['total_bill'] - tips['tip']
tips.head()
````

````python
# Gerando o gráfico de dispersão
sns.regplot(x = 'net_bill', y= 'tip', data= tips)
````

````python
# Gerando o gráfico de dispersão agora com a linha dos 10%
reg_1 = sns.regplot(x = 'net_bill', y= 'tip', data= tips)

# Gerando dados para a linha dos 10%
x = np.linspace(0,tips['net_bill'].max(),100)
y = .1*x

# Plotando a linha dos 10%
plt.plot(x, y, '-r', label='linha dos 10%')

# Opções estéticas para explicitar o (0,0) no gráfico
plt.xlim(left = 0)
plt.ylim(bottom = 0)
````

### 2. Mudança de perspectiva

Agora ajuste um modelo de regressão de ```tip_pct``` por ```net_bill```. Construa o gráfico de dispersão correspondente.

````python
# Gerando o gráfico de dispersão ed tip_pct por net_bill
reg_2 = sns.regplot(y = 'tip_pct', x= 'net_bill', data= tips[tips['tip_pct']<40])
reg_2
````

### 3. Comente

Compare os dois gráficos e comente (de forma objetiva e suscinta) a relação entre eles e a diferença do ajuste. Podemos comparar diretamente os $R^2$s?

````python
reg1 = smf.ols('tip ~ net_bill', data = tips).fit()
reg1.summary()
````

````python
reg2 = smf.ols('tip_pct ~ net_bill', data = tips).fit()
reg2.summary()
````

Podemos verificar que quanto maior o valor final da conta, maior a gorjeta, porém, ao mesmo tempo, quanto maior o valor da conta, menor a porcentagem de gorjeta.

O r2 é maior para o primeiro gráfico, porém, o que pode nos dizer que o primeiro gráfico possui uma relação mais forte que o segundo.

### 4. Ajuste do modelo em outra escala

Com o modelo no item 3, obtenha a preditiva do valor da gorjeta:
  1. obtenha o valor predito ```pred``` para ```tip_pct``` (dica, veja o atributo ```reg.fittedvalues```)
  2. obtenha o valor predito para ```tip``` como ```tip_pct * net_bill```
  3. calcule o $R^2$ do modelo para ```tip```
  4. compare esse $R^2$ com o do item 1

````python
# Para calcular o valor predito da pct de gorjeta
pred_tip_pct = reg2.fittedvalues
pred_tip_pct
````

````python
# Para obter o valor predito usaremos pred_tip= tip_pct * net_bill
pred_tip = pred_tip_pct * tips['net_bill']
pred_tip
````

````python
# Calculando o r-squared do modelo predito pred_tip
r2_score(tips['tip'], pred_tip)
````
