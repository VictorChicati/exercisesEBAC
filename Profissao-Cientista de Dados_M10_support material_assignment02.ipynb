1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
# Importando as bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# Lendo nossa base de dados
df = pd.read_csv('previsao_de_renda.csv')
df.head()
````

````python
# Separando apenas as colunas de interesse
df = df[['renda','posse_de_imovel','posse_de_veiculo']]
````

````python
# Comparando os valores
ax = sns.barplot(x="posse_de_veiculo", y="renda",
                   data=df, dodge=True, ci = 90)
````

````python
# comparando os valores
sns.barplot(x="posse_de_imovel", y="renda",
                   data=df, dodge=True, ci = 90)
````

Portanto, existem diferenças significantes na renda entre grupos com e sem posse de veiculo, assim como para grupos com e sem posse de imovel

````python
# Comparando todos os valores em apenas um gráfico
ax = sns.barplot(x="posse_de_veiculo", y="renda", hue="posse_de_imovel",
                   data=df, dodge=True, ci = 90)
ax.legend(bbox_to_anchor=(1, 1), loc=2, title='Posse de imovel')
````
